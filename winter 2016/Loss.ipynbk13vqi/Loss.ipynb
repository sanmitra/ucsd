```python
from __future__ import division
from array import array as pyarray
from numpy import append, array, int8, uint8, zeros, arange, vectorize, dot, random
from decimal import Decimal
import math
import os, struct

def read_mnist(images_file, labels_file, digits=arange(10)):
    flbl = open(labels_file, 'rb')
    magic_nr, size = struct.unpack(">II", flbl.read(8))
    lbl = pyarray("b", flbl.read())
    flbl.close()
    
    print [l for l in lbl if l == 0]
    fimg = open(images_file, 'rb')
    magic_nr, size, rows, cols = struct.unpack(">IIII", fimg.read(16))
    img = pyarray("B", fimg.read())
    fimg.close()

    ind = [ k for k in range(size) if lbl[k] in digits]
    N = len(ind)

    images = zeros((N, rows*cols), dtype=uint8)
    labels = zeros(N, dtype=int8)
    for i in range(len(ind)):
        images[i] = array(img[ ind[i]*rows*cols : (ind[i]+1)*rows*cols ])
        labels[i] = lbl[ind[i]]
    
    print [l for l in labels if l == 0]
    
    return images, labels

exp_fn = vectorize(math.exp)

def softmax_regression_gradient_descent(X, T, learning_rate=0.00000000001, threshold = 1):
    weights = [zeros(28*28)] * 10
    losses = []
    
    while True:
        loss = 0
        A = [exp_fn(dot(X, weights[i])) for i in range(10)]
        Y = A
        for n in range(len(X)):
            sum1 = 0
            for i in range(10):
                sum1 += A[i][n]
            
            for i in range(10):
                Y[i][n] = Y[i][n]/sum1
                
        for i in range(10):
            weights[i] = weights[i] + learning_rate * dot(X.transpose(), T[i]-Y[i])
        
        for n in range(len(X)):
            for i in range(10):
                loss += T[i][n] * math.log(Y[i][n])
        losses.append(loss)
        if len(losses) > 2 and math.fabs((losses[-1] - losses[-2])* 100/ losses[-2]) <= 1:
            break

    return weights
```

```python
# Training.
TRAIN_IMAGES  = "C:\\Users\\oop\\Desktop\\Winter 2016\\train-images.idx3-ubyte"
TRAIN_LABELS  = "C:\\Users\\oop\\Desktop\\Winter 2016\\train-labels.idx1-ubyte"
images_train, labels_train = read_mnist(TRAIN_IMAGES, TRAIN_LABELS)
#images_train, labels_train = images_train[:20000], labels_train[:20000]
```

```python
T = array([array([1 if label == i else 0 for label in labels_train]) for i in range(10)])
weights = softmax_regression_gradient_descent(images_train, T)
```

```python
# Testing.
TEST_IMAGES = "C:\\Users\\oop\\Desktop\\Winter 2016\\t10k-images-idx3-ubyte\\t10k-images.idx3-ubyte"
TEST_LABELS  = "C:\\Users\\oop\\Desktop\\Winter 2016\\t10k-labels-idx1-ubyte\\t10k-labels.idx1-ubyte"
images, labels = read_mnist(TEST_IMAGES, TEST_LABELS)
#images, labels = images[:2000], labels[:2000]

A = array([exp_fn(dot(images, weights[i])) for i in range(10)])
Y = A
for n in range(len(images)):
    sum1 = 0
    for i in range(10):
        sum1 += A[i][n]

    for i in range(10):
        Y[i][n] = Y[i][n]/sum1

# Report the overall test accuracy.
error_count = 0
for n in range(len(images)):
    probs = Y[:,n]
    #print probs
    predicted_digit = probs.argmax()
    if predicted_digit != labels[n]:
        #print  predicted_digit, labels[n]
        error_count += 1
print error_count
print "Overall test accuracy is " + str(100 - error_count * 100/ len(images))
```
